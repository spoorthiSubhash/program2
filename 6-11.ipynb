```python
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
```

```python
df.info()
```

```python
q1_w=df['Weight'].quantile(0.25)
q3_w=df['Weight'].quantile(0.75)
iqr=q3_w-q1_w
for i in df.index:
    if df.loc[i,'Weight']>(q3_w+1.5*iqr):
        df.drop(i,inplace=True)
    elif df.loc[i,'Weight']<(q1_w-1.5*iqr):
        df.drop(i,inplace=True)
```

```python
df.info()
```

```python
print(df.corr())
```

```python
x=df['Species']
x
```

```python
def Species_int(x):
    if x=='Bream':
        return 1
    if x=='Roach':
        return 2
    if x=='Whitefish':
        return 3
    if x=='Parkki':
        return 4
    if x=='Perch':
        return 5
    if x=='Pike':
        return 6
    if x=='smelt':
        return 7
```

```python
df['Species_num']=df['Species'].apply(Species_int)
df
```

```python
fig=sns.pairplot(df,hue='Species',diag_kind='hist')
fig.add_legend()
```

```python
x=df[['Weight','Length1','Length2','Length3','Height','Width']]
y=df['Species_num']
```

```python
lsvm.score(x_train,y_train)
```

```python
y_pred=lsvm.predict(x_test)
```

```python
##n-gram vectorizor 11b
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer 
```

```python
df=pd.read_excel('ngram_ex (1).xlsx')
df
```

```python
def get_ngrams(text,ngram_from=2,ngram_to=2,n=None,max_features=20000):
 vec=CountVectorizer(ngram_range=(ngram_from,ngram_to),
 max_features=max_features,stop_words='english').fit(text)
 bag_of_words=vec.transform(text)
 sum_words=bag_of_words.sum(axis=0)
 words_freq=[(word,sum_words[0,i])for word,i in vec.vocabulary_.items()]
 words_freq=sorted(words_freq,key=lambda x:x[1],reverse=True)
 return words_freq[:n]
```

```python
Unigrams=get_ngrams(df['product name'],ngram_from=1,ngram_to=1,n=15)
Unigrams_df=pd.DataFrame(Unigrams)
Unigrams_df.columns=['Unigram','Frequency']
Unigrams_df.head()
```

```python
bigrams=get_ngrams(df['product name'],ngram_from=2,ngram_to=2,n=15)
bigrams_df=pd.DataFrame(bigrams)
bigrams_df.columns=['bigram','Frequency']
bigrams_df.head()
```
